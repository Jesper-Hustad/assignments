```python
import pandas as pd

data = pd.read_csv('agaricus-lepiota.data')
```

Distrbution in data:

```python
data[['habitat', 'edibility']].groupby('habitat').describe().plot.bar()
```

Convert each feature to dummy variables:

```python
dummies = pd.get_dummies(data)
dummies
```

Visualise the feature space:

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(100,150)) 
fig = plt.gcf()
plt.spy(dummies, markersize=0.8) 
plt.show()
```
